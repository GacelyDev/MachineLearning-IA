## Missing values

```python
#Comprobar si estan los valore en el dataset son nulos brindando una columna como parametro
pd.isnull(data["body"])
```

```python
#Comprobar si estan los valore en el dataset no son nulos brindando una columna como parametro
#ravel construye un unico array de datos
pd.notnull(data["body"])
```

```python
#ravel construye un unico array de datos
#determina la cantidad de valores nulos que hay
pd.isnull(data["body"]).values.ravel().sum()
```

```python
#ravel construye un unico array de datos
#determina la cantidad de valores no nulos que hay
pd.notnull(data["body"]).values.ravel().sum()
```

Los valores que faltan en un dataset pueden darse por dos razones:

```python
#Función general para añadir y eliminar columnas con sus respectivos valores

def createDummies(df, var_name):
    dummy = pd.get_dummies(data[var_name], prefix=var_name)
    df = df.drop(var_name, axis =1)
    df = pd.concat([df,dummy], axis=1)
    return df
```
